<b>What are missing values?</b>
<p>Missing values occur when data is unavailable or not recorded.</p>

<p>Many models cannot handle nulls, Biases results, Reduces prediction accuracy</p>

<h4>Handling Missing Values -- 3 approaches</h4> 

<h5>Method 1 Remove Missing Data (deletion)</h5>
<p>it's mostly used when dataset is large and the column isn't important </p>

```python
df.dropna(inplace=True) #drop rows with null values and inplace is used to alter the data without assigning a new variable
```

```python
m2 = df.dropna(subset=['urgency_level']) #Drop specific column null rows
# m2.isnull().sum()
```

```python
m3 = df.drop(columns=["belief_alignment_status"]) # Drop columns with many nulls
# m3.isnull().sum() 
```

<h5>Method 2 — Fill Missing Values (Imputation)</h5>

<p><b>df['response_character_count'].fillna('response_character_count').mean()</b> --> filling with mean(numerical data)</p> 

<p><b>df['response_character_count'].fillna(df['response_character_count'].median(), inplace=True)</b> --> Fill with median (better if outliers exist)</p> 

```python
# Fill with constant value

fm3 = df.fillna(0)
# fm3.isnull().count()

fm4 = df.fillna("Unknown")
# fm4.isnull().count()
```

```python
# Fill with mode (categorical data)
df["urgency_level"].fillna(df["urgency_level"].mode()[0])
```

```python
df["belief_alignment_status"].fillna(df["belief_alignment_status"].mode()[0])
```

```python
df["fact_check_method_used"].fillna(df["fact_check_method_used"].mode()[0])
```
